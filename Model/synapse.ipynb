```python
# setup llm & embedding model
llm=Ollama(model="llama3.1", request_timeout=120.0)
# embed_model = HuggingFaceEmbedding( model_name="Snowflake/snowflake-arctic-embed-m", trust_remote_code=True)
embed_model = HuggingFaceEmbedding( model_name="BAAI/bge-large-en-v1.5", trust_remote_code=True)
```

```python
def load_data(input_dir_path):
  loader = SimpleDirectoryReader(
            input_dir = input_dir_path,
            required_exts=[".pdf"],
            recursive=True
        )
  docs = loader.load_data()

  # Creating an index over loaded data
  Settings.embed_model = embed_model
  index = VectorStoreIndex.from_documents(docs, show_progress=True)
  return index
```

```python
def format_streaming_response(response):
    # Check if the response is of type StreamingResponse
    if hasattr(response, 'source_nodes') and isinstance(response.source_nodes, list):
        formatted_output = []

        for node_with_score in response.source_nodes:
            node = node_with_score.node
            # Extract the text and metadata
            text = node.text
            metadata = node.metadata
            # Extract filename and page label from metadata
            file_name = metadata.get('file_name', 'Unknown file')
            page_label = metadata.get('page_label', 'Unknown page')

            # Format the metadata for better readability
            metadata_str = "\n".join(f"{key}: {value}" for key, value in metadata.items())

            # Combine metadata, file info, and text
            formatted_node = (
                f"Source File: {file_name}\n"
                f"Page: {page_label}\n\n"
                f"Metadata:\n{metadata_str}\n\n"
                f"Content:\n{text}\n\n"
            )
            formatted_output.append(formatted_node)

        # Join all formatted nodes
        return "\n".join(formatted_output)

    return "No relevant information found in the response."
```

```python
def manifestor_comparator():
    # Initialize key candidates and response accumulator
    keys = ["AKD", "DJ", "NR", "SP"]
    final_response = ""

    # Greet the user
    print("🌟 Welcome to the Manifestor Comparator! 🌟")
    user_input = input("🔍 Please enter your question: ")

    user_input = model.generate_content(f"Using {user_input} understand what are about his question").text

    # Process the query for each candidate
    for key in keys:
        query_engine = dir[key].as_query_engine(streaming=True, similarity_top_k=4)
        response = query_engine.query(user_input)
        formatted_response = format_streaming_response(response)
        final_response += f"\n{formatted_response}"

    # Generate content based on the collated responses
    gem_res = model.generate_content(
        f"{final_response}\n\nUsing the gathered information, generate a table with:"
        "\n **Candidate Name** (Anura Kumara Dissanayake, Sajith Premadasa, Namal Rajapaksha, Dilith Jayaweera),"
        "\n **Their Plans** (clearly and well-detailed),"
        "\n **Comparative Analysis** of their policies, showing how they differ and their potential impacts."
    )

    # Return the generated comparison table in markdown format
    return to_markdown(gem_res.text)
```

```python
# Initialize conversation history
conversation_history = []

def update_conversation_history(role, content):
    """Append a new exchange to the conversation history."""
    conversation_history.append({"role": role, "content": content})
```

```python
def get_prompt_from_history():
    """Create a prompt from the conversation history."""
    prompt = ""
    for turn in conversation_history:
        prompt += f"{turn['role']}: {turn['content']}\n"
    return prompt
```

```python


def get_response_for_election(query):
    # Update conversation history with the new user input
    update_conversation_history("user", query)

    # Create a prompt including conversation history
    prompt = get_prompt_from_history()
    simple_promt = model.generate_content(prompt + "Simply extract what exaclty user want to know from this input")

    # Query the engine with the prompt
    query_engine = elec_data.as_query_engine(streaming=True, similarity_top_k=4)
    response = query_engine.query(simple_promt.text)

    # Format the response from the query engine
    formatted_response = format_streaming_response(response)

    # Generate a response using the model
    GEMRES = model.generate_content(f"Read this {formatted_response} and give a good answer to the {prompt} (no need to mention page numbers)")

    # Update conversation history with the AI response
    ai_response = GEMRES.text
    update_conversation_history("AI", ai_response)

    # Return the AI response in markdown format
    return to_markdown(ai_response)



```

```python
from IPython.display import Markdown, display
def start_chat():
    print("Chatbot: Hello! How can I assist you today?")

    while True:
        # Get user input
        user_input = input("You: ")

        # Exit the chat loop if the user types 'exit'
        if user_input.lower() in ['exit', 'quit', 'bye']:
            print("Chatbot: Goodbye! Have a great day!")
            break

        # Get response from the chatbot
        response = get_response_for_election(user_input)

        # Print the AI response
        display(response)
```

```python

# Call the function to start the chat
start_chat()
```
